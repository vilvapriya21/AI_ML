Terminology :

Token       = smallest unit of text AI reads ("Hello" = 1 token)

Context     = entire conversation AI remembers

Embedding   = text converted to numbers (vectors)

Inference   = AI generating a response

Parameters  = model's brain size (7B, 13B)

Prompt      = your input to the AI

Completion  = AI's output

System Role = instructions that control AI behavior

LLM Parameters

```python

payload = {
    "model": "mistral",
    "prompt": "Your question here",
    
    # CREATIVITY CONTROL
    "temperature": 0.7,      # 0.0 = robotic/exact, 2.0 = very creative/random
    
    # LENGTH CONTROL  
    "num_predict": 200,      # max words (tokens) in response
    
    # FOCUS CONTROL
    "top_p": 0.9,            # 0-1, lower = more focused word choices
    "top_k": 40,             # how many words AI considers at each step
    
    # REPETITION CONTROL
    "repeat_penalty": 1.1,   # >1 = avoids repeating same words
    
    "stream": False
}
```
