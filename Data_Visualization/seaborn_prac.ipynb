```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

import os
print(os.getcwd())
df = pd.read_csv("../Datasets/titanic.csv")
#df=pd.read_csv("C:/Users/Priya Koma/Desktop/AI_ML/Datasets/titanic.csv")
df.info()
```
