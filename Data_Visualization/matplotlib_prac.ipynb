```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 15, 50, 13]

plt.plot(x, y) #plot: to show trends or changing data like stocks 
#labels
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("My First Plot")
plt.show()
```

```python
#bar plot : to compare categories

subjects=['Math','Science','English']
scores=[80,70,90]

plt.bar(subjects,scores)
```

```python
ages=[5, 7, 8, 12, 15, 16, 18, 20, 22, 25]#for distribution

plt.hist(ages, bins = 5)
plt.xlabel("Age")
plt.ylabel("Number of People")
plt.title("Age Distribution")
plt.show()
```
